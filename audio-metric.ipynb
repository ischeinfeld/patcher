# Audio Similarity Metrics

## Prepare Audio

```python
y, sr = librosa.load(librosa.example('vibeace', hq=True), sr=44100)
fig, ax = plt.subplots(figsize=(14, 5))
librosa.display.waveshow(y, sr=sr, ax=ax)
```

```python
Audio(y[:10 * sr], rate=sr)
```

```python
# Make a new Pedalboard object that contains plugins
board = pb.Pedalboard([
    pb.Phaser(rate_hz=1.0, depth=1.0, feedback=0.25, mix=1.0),
    pb.LadderFilter(mode=LadderFilter.HPF12, cutoff_hz=329.63, resonance=0.75),
    pb.Gain(-5),
    pb.Reverb(room_size=0.75, wet_level=0.5),
    pb.Limiter(),
], sample_rate=sr)

effected = board(y, sample_rate=sr)
```

```python
y, sr = librosa.load(librosa.example('vibeace', hq=True), sr=44100)
fig, ax = plt.subplots(figsize=(14, 5))
librosa.display.waveshow(effected, sr=sr, ax=ax)
```

```python
Audio(effected[:20 * sr], rate=sr)
```
